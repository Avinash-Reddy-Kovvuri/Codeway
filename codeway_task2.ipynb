```python
#seperating the data for analysis
legit = dataset[dataset.is_fraud==0]
fraud = dataset[dataset.is_fraud==1]
print(fraud.shape)
```

```python
print(legit.shape)
```

```python
#statistical measures of data
dataset.groupby('is_fraud').mean()
```

```python
new_dataset = pd.concat([legit,fraud],axis=0)
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['is_fraud'].value_counts()
```

```python
new_dataset = new_dataset.drop(['Unnamed: 0', 'trans_date_trans_time', 'cc_num', 'trans_num'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
categorical_columns = ['merchant', 'category', 'first', 'last', 'gender', 'street', 'city', 'state', 'job', 'dob']
for col in categorical_columns:
    new_dataset[col] = label_encoder.fit_transform(new_dataset[col])
new_dataset.head()
```

```python
x = new_dataset.drop(columns='is_fraud',axis=1)
y=new_dataset['is_fraud']
```

```python
x
```

```python
new_dataset.info()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify = y ,random_state=1)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_train.head()
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
x_pred = model.predict(x_train)
train_data_accuracy = accuracy_score(x_pred,y_train)
```

```python
print("traing_data_accuracy : ",train_data_accuracy)
```

```python
y_pred = model.predict(x_test)
test_data_accuracy = accuracy_score(y_pred,y_test)
```

```python
print("testing_data_accuracy : ",test_data_accuracy)
```
